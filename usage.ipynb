# Library Usage

```python
import charsec
```

## Encoding and decoding data

We can encode a secret message in a space character like so:

```python
space = charsec.encode(" ", "Secret Message")
space
```

Then insert the modified space character with the data into the message:

```python
message = f"Hello{space}world!"
message
```

Then decode it:

```python
charsec.decode(message)
```

## Checking if text contains hidden data

```python
charsec.check(message)
```

## Executing hidden code

```python
hidden_code = "print('Hello world!')"
space = charsec.encode(" ", hidden_code)
code = f"#{space}You would think this wouldn't run, but it does!"
code
```

```python
charsec.exec(code, debug=True)
```

## Running hidden code in a file

```python
charsec.run("demo/simple.py", debug=True)
```

## Hiding/recovering files in/from other files

```python
! uv pip install pillow
```

We can encode the data like so:

```python
with open("demo/meme.jpeg", "rb") as jpeg:
    data = jpeg.read()
space = charsec.encode(" ", data)
message = f"lorem{space}ipsum"
message
```

And recover it like so:

```python
from PIL.Image import open
from io import BytesIO
```

```python
im = open(BytesIO(charsec.decode(message)))
im
```

But also, the inject and extract functions are available to simplify things.

```python
# this may raise an exception, currently you cannot add hidden data to a file already containing hidden data
charsec.inject("demo/meme.jpeg", "demo/test.md")
```

```python
charsec.extract("demo/test.md", "demo/hidden.jpeg")
```
